### Zadanie 2

```python
csv = pd.read_csv("dane_sprzedazy.csv") #sczytywanie danych z pliku .csv
```

### Zadanie 3

```python
csv.head() #wyświetlenie pierwszych pięciu wierszy
```

```python
csv.tail() #wyświetlenie ostatnich pięciu wierszy
```

```python
csv.describe() #wyświetlenie statystyk dla kolumn typów numerycznych
```

### Zadanie 4

```python
csv_filtered = csv.loc[csv["cena"] > 10, ["cena", "ilość"]] #filtrowanie danych, najpierw podajemy warunki filtrowania, a potem kolumny, z których dane chcemy przefiltrować
csv_filtered
```

### Zadanie 5

```python
new_csv = csv #utworzenie kopii DataFrame
new_csv["wartość"] = new_csv["cena"] * new_csv["ilość"] #dodanie nowej kolumny na końcu DataFrame, dane w tej kolumnie zależne są od danych w kolumnach "cena" oraz "ilość"
new_csv
```

### Zadanie 6

```python
new_csv.groupby("nazwa").sum("wartość") #grupowanie danych według nazwy, suma wartości poszczególnych produktów
```

### Zadanie 7

```python
new_csv.sort_values("wartość", ascending=False) #sortowanie malejące, według wartości każdego produktu
```

### Zadanie 8

```python
new_csv.loc[2:4, ["nazwa", "wartość"]] #wycięcie danych z DataFrame, tylko kolumny "nazwa" oraz "wartość", wiersze 2 - 4
```
