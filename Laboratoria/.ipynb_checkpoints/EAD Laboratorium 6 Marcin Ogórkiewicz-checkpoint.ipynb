### Zadanie 2: Analiza korelacji
Korzystając z wyników raportu wygenerowanego w zadaniu 1, zidentyfi kuj kolumny, które są silnie skorelowane (wartość korelacji powyżej 0.8 lub poniżej -0.8). Wyświetl dane tylko dla tych kolumn i sprawdź, czy ich usunięcie wpłynęłoby na analizy.



```python
def normalisation(x, xmin, xmax): #definicja funkcji będącej reprezentacją wzoru
    x=(x-xmin)/(xmax-xmin)
    return x

new_price = csv['price'].apply(lambda x: normalisation(x, csv['price'].min(), csv['price'].max())) #podstawienie wartości kolumny price do funkcji normalizacji
price_DataFrame = pd.DataFrame({  #tworzenie DF do porównania i histogramu
    'price': csv['price'],        #oryginalne ceny
    'new_price': new_price        #znormalizowane ceny
})
```

```python
price_DataFrame['price'].plot(kind='hist', bins=10, color='green', edgecolor='black', title='Histogram wartości kolumny price') #tworzenie histogramu
plt.xlabel('Wartości')
plt.ylabel('Częstotliwość')
plt.show()
```

### Zadanie 3: Rozbudowany raport
Wygeneruj bardziej zaawansowany raport, zmieniając ustawienia pandas-profiling, aby:
➔ Włączyć analizę wartości odstających (outliers).
➔ Zwiększyć dokładność analizy liczbowej poprzez dostosowanie liczby próbek do większego zestawu danych.
Wyświetl najciekawsze wykryte cechy danych (np. nietypowe rozkłady, kolumny z małą ilością unikalnych wartości).



```python
def standarisation(x, xavg, xstd): #definicja funkcji będącej reprezentacją wzoru
    z = (x-xavg)/xstd
    return z

new_salary = csv['salary'].apply(lambda x: standarisation(x, csv['salary'].mean(), csv['salary'].std())) #podstawienie wartości kolumny price do funkcji normalizacji
salary_DataFrame = pd.DataFrame({  #tworzenie DF do porównania i histogramu
    'salary': csv['salary'],        #oryginalne pensje
    'new_salary': new_salary        #zestandaryzowane pensje
})
```

```python
salary_DataFrame['salary'].plot(kind='hist', bins=10, color='green', edgecolor='black', title='Histogram wartości kolumny salary') #tworzenie histogramu
plt.xlabel('Wartości')
plt.ylabel('Częstotliwość')
plt.show()
```
